```python
from TernaryPlot import * 
import matplotlib.pyplot as plt
```

```python
BS  = clean_briefsummary(load_briefsummary())
```

```python
components  = ['Co','W','Cr']
```

```python
BS[components] = get_fractions_by_components(BS,components)
```

```python
plt.rc('axes', labelsize=32, labelpad=10 )
plt.rc('legend',fontsize=24)
mplfig = ternary_matplotlib(BS, components)
ax = mplfig.get_axes()
ax[0].tick_params(labelsize=30,pad=10, width=1, length=8)
ax[0].set_title('')
leg = ax[0].legend()
leg.set_bbox_to_anchor((0.95,1.1))
for lh in leg.legendHandles:
    lh.set_sizes ([100])
    lh.set_alpha(1)
mplfig.tight_layout()
mplfig.savefig('ternary_plot.pdf')
```
