```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('CrWCo.csv')
df
plt.plot(df['composition'], df['D1'], 's', color='white', markeredgecolor='black')
plt.plot(df['composition'], df['D2'], 's', color='white', markeredgecolor='black')
plt.plot(df['composition'], df['C1'], 's', color='white', markeredgecolor='green')
plt.plot(df['composition'], df['B1'], 's', color='white', markeredgecolor='blue')
plt.plot(df['composition'], df['B2'], '*', color='white', markeredgecolor='blue')
for i in range(6):
    plt.plot(df['composition'], df[f'A{i+1}'], '<', color='white', markeredgecolor='red')

plt.ylim(-0.02, 1.02)
plt.xlim(0.26, 0.4)
plt.xlabel('at. % Mo/W atom')
plt.ylabel('Site occupancy')
```
