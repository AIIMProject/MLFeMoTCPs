# a little look to the energies

```python
from mendeleev import element as mendeleev_elements 
import re
import numpy as np
```

```python
for symbol in elements:
    if symbol != '':
        thissym = re.sub('_[a-z]+','',symbol)
        atom_volume = (4/3)*np.pi*(mendeleev_elements(thissym).atomic_radius/100)**3
        atom_name = mendeleev_elements(thissym).name
        print('the volume of {} is {:.2f}'.format(atom_name, atom_volume))
        
```
